### PCA

**Extração de Características:** Encontra os relacionamentos dos melhores atributos e cria novos atributos.

É um algoritmo de aprendizagem não supervisionada.

Aplica-se em dados linearmente separáveis.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
df = df.drop(labels = 'Weight', axis = 1)
```

```python
df['Preferred Foot'].replace({'Left': 0, 'Right': 1}, inplace=True)
```

```python
pca_transform = pca.fit_transform(df)
```

```python
pca_transform.shape
```

```python
pca_transform
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
import plotly.express as px
```

```python
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    pca_transform,
    labels=labels,
    dimensions=range(2), 
    color=df["Value"]
)
fig.update_traces(diagonal_visible=False)
fig.show()
```

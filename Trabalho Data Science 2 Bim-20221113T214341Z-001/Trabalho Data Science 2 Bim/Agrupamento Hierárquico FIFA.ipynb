## **Agrupamento Hierárquico**


```python
df_esc
```

```python
df_esc.shape
```

**Dendograma**

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hier = AgglomerativeClustering(n_clusters=7, affinity='euclidean', linkage = 'average')
classificacao = hier.fit_predict(df_esc)
```

```python
classificacao
```

A visualização fica interessante quando tempos de dois a tres dimensões. Neste caso estamos trabalhando com 4 dimensões. 


**Agrupamento**

```python
agrupamento = pd.DataFrame(classificacao, columns = ['ranking geral'])
agrupamento
```

```python
df_hc = pd.concat([df, agrupamento],axis=1)
df_hc
```
